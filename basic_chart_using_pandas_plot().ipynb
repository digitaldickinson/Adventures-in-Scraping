## Cleaning up empty cells

That looks good, 50 rows of population data.  But we actually have two problems in the data as we've loaded it. 

The first is a weird NaN on row 58. NaN means 'Not a number'. At this point, there isn't a problem per se. It's just that we have no data for 2018. But programming languages are quite particular with types of data. When we come to plot the file, the mix of numbers and the NaN can sometimes 

We should really have 'cleaned' that up before saving the data. But we may not always have that chance and it's good to have a way to check the data we have for any weird values. This is where a specialist library like pandas comes into its own. 

Run the following code. 

```
data.isnull()
```

The code here is pretty easy to decipher.  

* `data.isnull()` - takes the data and runs the function `isnull()`which detects any null (or NaN) values. It returns a True or False response; `False` if there is a number and `True `if there is no value. You can see when we run the code that Row 58 has `True` in the `Population `Column but false everywhere else. 

* `.values.any() `

In some circumstances we might replace the value with a number (e.g. 0) to show no date but that could be confusing. In this case we can use another simple function to get rid of the row all together.


```
newdata = data.dropna()

data = newdata

data
```

We can see the data has gone, and the table finishes at 2017.  Again, the code is pretty simple

* `newdata = data.dropna()` - creates a new dataframe called newdata to temporarily store the corrected data

* `data = newdata` - now the data has been edited I can 'overwrite the old variable by copying in the contents of newdata

I could have done the same thing with 

* `data =  data.dropna()`

But there's another problem.  The commas in the population figured, whilst helpful in terms of readability, are going to be a problem.  

Even though we see them as numbers, because of the commas, Pandas doesn't.  We can see that if we run a command called `.ftypes` to show us what type of data we have in our dataframe

```
data.ftypes
```

The year is a number - int64 means an integer or whole number.The population should be the same. But its actually an object - panda's name for a string of text.

We need to convert this into a number. 

```
data["Population"] = data["Population"].str.replace(",","").astype(int)
```

This code here looks a bit complicated than the code we used before, but it follows the same principles. Breaking things down:

*  `data["Population"] =` This means we just want to store any results in the `Population` column in `data` 

*  `data["Population"].str.replace(",","")` - this means take whatever is the Population column already, and replace any text (called a string or str here) with whatever we define. In this case we want to replace commas with spaces `(",","")`. This means `65,595,565` would become `65 595 565` 

The data is still text (or String) so we need to correct that. 

*  `.astype(int)` - this means anything that gets changed is saved as a whole number (integer)

Now if we look at the types of data, we can see both columns are now integers. 


```
data.ftypes
```

And if we look at the contents of the data, we can see the commas have gone

```
data
```

## Plotting the data

Now the data is clean and in the right format, we can plot a chart. Pandas has some basic charts built in, we just need to tell it to `plot() `the columns. 


```
plt = data.plot(x='Year', y='Population')
```

The code here is pretty simple

* `plt =` creates a new object to store our plotted data in
* `data.plot(x='Year', y='Population')` - takes the data and calls the `plot()` function. We can define the x and y axis of the plot as variables. `x `is the values from the` Year `column. `y`, the `Population` column

## Saving an Image 

Looks pretty good, even straight out of the box.  All we need to do now is save it an image we can use. 

The code below grabs our plot and wraps everything up together as a  `figure` called fig. 

Once we've done that we can use a function called `savefig()` to spit out a jpg file. 

```
fig = plt.get_figure()
fig.savefig('figure.jpg')
```
